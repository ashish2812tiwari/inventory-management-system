```python
c = [50,100,200,90,50]
d = [50,100,200,90,50]
b = [1,2,3,4,5]
f = ["Laptop","Desktop","Keyboards","Mother board","RAM"]
print("%35s  "%("Items present in store"))
print("----------------------------------------------------------------------")
print("%10s %10s %20s %20s"%("Product code","Item Name","Quantity available","Quantity can be issued"))
print("%7d %13s %20d %10d"%(b[0],f[0],c[0],d[0]))
print("%7d %13s %20d %10d"%(b[1],f[1],c[1],d[1]))
print("%7d %13s %20d %10d"%(b[2],f[2],c[2],d[2]))
print("%7d %13s %20d %10d"%(b[3],f[3],c[3],d[3]))
print("%7d %13s %20d %10d"%(b[4],f[4],c[4],d[4]))
print('Press \n\n 1 : To update the store\n 2 : To Issue the items')
a=int(input())
if a==1:
    while True:
        choice = int(input("Enter product code number to Update : "))
        if choice in b:
            if choice == 1:
                print("\n%35s"%("Laptop"))
                num = int(input("Enter the quantity : "))
                cc=c[0]+num
                c[0]=cc
                d[0]=cc
                
            elif choice == 2:
                print("\n%35s"%("Desktop"))
                num = int(input("Enter the quantity : "))
                ccc=c[1]+num
                c[1]=ccc
                d[1]=ccc
                
            elif choice == 3:
                print("\n%35s"%("Keyboards"))
                num = int(input("Enter the quantity : "))
                cc=c[2]+num
                c[2]=cc
                d[2]=cc
                
            elif choice == 4:
                print("\n%35s"%("Mother board"))
                num = int(input("Enter the quantity : "))
                cc=c[3]+num
                c[3]=cc
                d[3]=cc
                
            elif choice == 5:
                print("\n%35s"%("RAM"))
                num = int(input("Enter the quantity : "))
                cc=c[4]+num
                c[4]=cc
                d[4]=cc
                
            exit = input("\nDo you want to add new item or to exit? \nPress \n1. To continue \n2. To exit: ")
            if exit=='2':
                print("\n\n%35s  "%("Thank You\n\n"))
                print("%40s  "%("Updated store"))
                print("-----------------------------------------------------------")
                print("%10s %10s %20s %20s"%("Product code","Item Name","Quantity available","Quantity can be issued"))
                print("%7d %13s %20d %10d"%(b[0],f[0],c[0],d[0]))
                print("%7d %13s %20d %10d"%(b[1],f[1],c[1],d[1]))
                print("%7d %13s %20d %10d"%(b[2],f[2],c[2],d[2]))
                print("%7d %13s %20d %10d"%(b[3],f[3],c[3],d[3]))
                print("%7d %13s %20d %10d"%(b[4],f[4],c[4],d[4]))
                break
        else :
            print("\n%35s"%("Enter valid product code to update"))
                
#---------------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------------
elif a==2:
    e=[]
    while True:
        choice = int(input("Enter product code number to Issue : "))
        if choice in b:
            if choice == 1:
                print("\n%35s"%("Laptop"))
                num = abs(int(input("Enter the quantity : ")))
                if num<=c[0] and num>0:
                    e.append(choice)
                    cc = c[0] - num
                    c[0]=cc
                    d[0]=num
                else:
                    print('%20s'%'Currently item out of stock')
                    print('%20s %5d'%('Quantity availabe is',c[0]))
                
            elif choice == 2:
                print("\n%35s"%("Desktop"))
                num = abs(int(input("Enter the quantity : ")))
                if num<=c[1] and num>0:
                    e.append(choice)
                    cc = c[1] - num
                    c[1]=cc
                    d[1]=num
                else:
                    print('%20s'%'Currently item out of stock')
                    print('%20s %5d'%('Quantity availabe is',c[1]))
                
            elif choice == 3:
                print("\n%35s"%("Keyboards"))
                num = abs(int(input("Enter the quantity : ")))
                if num<=c[2] and num>0:
                    e.append(choice)
                    cc = c[2] - num
                    c[2]=cc
                    d[2]=num
                else:
                    print('%20s'%'Currently item out of stock')
                    print('%20s %5d'%('Quantity availabe is',c[2]))
                
            elif choice == 4:
                print("\n%35s"%("Mother board"))
                num = abs(int(input("Enter the quantity : ")))
                if num<=c[3] and num>0:
                    e.append(choice)
                    cc = c[3] - num
                    c[3]=cc
                    d[3]=num
                else:
                    print('\n%20s'%'Currently item out of stock')
                    print('%20s %5d'%('Quantity availabe is',c[3]))
                
            elif choice == 5:
                print("\n%35s"%("RAM"))
                num = abs(int(input("Enter the quantity : ")))
                if num<=c[4] and num>0:
                    e.append(choice)
                    cc = c[4] - num
                    c[4]=cc
                    d[4]=num
                else:
                    print('%20s'%'Currently item out of stock')
                    print('%20s %5d'%('Quantity availabe is',c[4]))
                
            exit = input("\nDo you want to Issue more item or to exit? \nPress \n1. To continue \n2. To exit: ")
            if exit=='2':
                print("\n\n%35s  "%("Thank You\n\n"))
                print("%40s  "%("Issued material store"))
                print("-----------------------------------------------------------")
                print("%10s %10s %20s %20s"%("Product code","Item Name","Quantity available","Quantity can be issued"))
                for i in range(len(e)):
                    if choice in e:
                        r = e[i]-1
                        print("%7d %13s %20d %10d"%(b[r],f[r],c[r],d[r]))
                    else:
                        print('invalid input')
                break
        else :
            print("\n%35s"%("Enter valid product code to update"))
    
else:
    print('Please enter valid input')
    
```
